```python
students.loc[6,'이름' : '수학'] = ['별님',50,60] #이름부터 수학까지 순서대로
```

```python
students
```

```python
students.loc[(students['국어'] >=80)&(students['수학']>=80),'합격'] = 'pass'
```

```python
students.loc[students['합격']!='pass','합격'] = 'fail' # NOT 합격인 애들 중 합격 칼럼에 FAIL을 추가한다.
```

```python
students
```

```python
condition_list = [(students['국어'] >= 90),
                 (students['국어'] >= 80)&(students['국어'] < 90),
                 (students['국어'] >= 70)&(students['국어'] < 80)]
choice_list = ['A','B','C']
students['점수'] = np.select(condition_list,choice_list,default = "F") #디폴트값을 설정하는 이유는 해당 상태값에 포함이 안되면 F를 주기 위함.
```

# Merge

```python
product = pd.DataFrame({'상품코드' : ['G1','G2','G3','G4'],
                       '상품명' : ['우유', '감자', '빵', '치킨'],})
sale = pd.DataFrame({'주문번호' : [1001,1002,1002,1003,1004],
                    '상품코드' : ['G4','G3','G1','G3','G5'],
                    '주문수량' : [1,4,2,2,3]})
sale.merge(product, on = '상품코드', how = 'inner')
```

```python
sale.merge(product, on = '상품코드', how = 'outer')
```

```python
sale.merge(product, left_on = '상품코드', right_on = '상품코드', how = 'left')
```

```python
product = pd.DataFrame({'상품코드' : ['G1','G2','G3','G4'],
                       '상품명' : ['우유', '감자', '빵', '치킨'],})
sale = pd.DataFrame({'주문번호' : [1001,1002,1002,1003,1004],
                    '상품코드' : ['G4','G3','G1','G3','G5'],
                    '주문수량' : [1,4,2,2,3]})
merged = sale.merge(product, left_on = '상품코드', right_on = '상품코드', how = 'left')
print(merged)
```
